#  Language modeling

## bigram
1. select pairs

```python
bigram_model = defaultdict(lambda: defaultdict(lambda: 0))
for sentance in en_sentences:
    words=sentance.split(' ')
    len_words=len(words)
    for i in range(0,len_words-3):
        w1,w2 = words[i:i+2]
        bigram_model[w1][w2] += 1
```

2. calculate probability

```python
for w1 in bigram_model:
    total = sum(bigram_model[w1].values())
    for w2 in bigram_model[w1]:
        bigram_model[w1][w2] /= total
```

Find long sentaance

```python
long_sentances = list(filter(lambda sentance: len(sentance.split(' ')) > 100, en_sentences))
```

probability of the whole sequence

```python
def probability_sentances(sentances):
    probability_sentances_array=[]
    for sentance in sentances:
        words=sentance.split(' ')
        len_words=len(words)
        multiple_probability = 1
        for i in range(0,len_words-3):
            w1,w2 = words[i:i+2]
            multiple_probability *= bigram_model[w1][w2]
        probability_sentances_array.append((len_words,multiple_probability))
    return probability_sentances_array
```

## Translation modeling

```python
def prob_all_alignments(source_sentence, target_sentence):#source = swed
    alignment_prob_for_t = defaultdict(lambda: 0.0)
    for target in target_sentence:
        for source in source_sentence:
            alignment_prob_for_t[target] += probability[(target,source)]
    return alignment_prob_for_t
```

```python
count = {}
target_total={}
for swed_word in swed_words:
    for eng_word in eng_words:
        count[(swed_word,eng_word)] = 0
        target_total[eng_word]=0

```

```python
for aligned_sentence in parallel:
    src_sentence = aligned_sentence[1]
    trg_sentence = aligned_sentence[0] 
    total_count = prob_all_alignments(src_sentence, trg_sentence)
    for t in trg_sentence:
        for s in src_sentence:
            counts = probability[(t,s)]
            normalized_count = counts / total_count[t]
            count[(t,s)]=0
            target_total[s]=0
            count[(t,s)] += normalized_count
            target_total[s] += normalized_count
```

Recalculating probability

```python
for swed_word in swed_words:
    for eng_word in eng_words:
        probability[(swed_word,eng_word)] = count[(swed_word,eng_word)]/target_total[eng_word]
```

```python
alignments = []
for i in range(len(swed_words)):
    print("Alignment for sentence {0}:".format(i+1))
    align = []
    for word in swed_words[i]:
        scores = [probability[(word, ew)] for ew in eng_words[i]]
        ind = scores.index(max(scores))
        align.append((swed_words[i].index(word), ind))
    alignments.append(align)
    print(align)
```
